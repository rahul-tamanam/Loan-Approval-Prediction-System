### Outlier Removal

```python
#remove outliers using IQR method
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
outliers1 = ((data >= (Q3 + 1.5 * IQR))).sum()
print("High Outliers in each column:\n", outliers1)
```

```python
outliers2 = (( data <= (Q1 - 1.5 * IQR))).sum()
print("Low Outliers in each column:\n", outliers2)
```

```python
# Filter out rows with outliers
no_outliers = ~(data >= (Q3 + 1.5 * IQR)).any(axis=1)
fd = data[no_outliers]
fd
```

```python
# Reset index after filtering out rows
data.reset_index(drop=True, inplace=True)
```

### Model Training and Validation

```python
# Function for training and validating models
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

model_df = {}

def model_val(model, X, y):

    # Split the data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.30, random_state=21)

    # Convert data to float
    X_train = np.array(X_train).astype("float32")
    X_test = np.array(X_test).astype("float32")

    # Convert labels to float (as float is more robust)
    y_train = np.array(y_train).astype("float32")
    y_test = np.array(y_test).astype("float32")

    # Train the model with the training data
    model.fit(X_train, y_train)

    # Make predictions on the test data
    y_pred = model.predict(X_test)

    # Calculate and print the accuracy score
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model} accuracy is {np.mean(accuracy)}")

    # Calculate and print the average cross-validation score
    score = cross_val_score(model, X, y, cv=5)
    print(f"{model} Avg cross val score is {np.mean(score)}")

    # Calculate and print the classification report of the model
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    # Add the model's cross-validation score to model_df
    model_df[model] = round(np.mean(score) * 100, 2)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Assuming X and y are already defined as your features and target variables

# Define models
models = {
    'Logistic Regression': LogisticRegression(),
    'Support Vector Machine': svm.SVC(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Random Forest (Custom)': RandomForestClassifier(n_estimators=25, min_samples_split=2, max_depth=30, max_features=1)
}

# Function to evaluate model using cross-validation
def model_val(model, X, y):
    scores = cross_val_score(model, X, y, cv=5)  # 5-fold cross-validation
    return scores.mean() * 100  # Return the mean of the scores as a percentage

# Store the cross-validation scores
model_scores = {name: model_val(model, X, y) for name, model in models.items()}

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(model_scores.keys(), model_scores.values(), color='skyblue')
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('Model Cross-Validation Accuracy Comparison')
plt.ylim(0, 100)  # Ensure the y-axis is from 0 to 100
plt.xticks(rotation=45, ha='right')  # Rotate model names for better readability
plt.tight_layout()
plt.show()
```

### Neural Network Implementation

```python
# Define a neural network model using Keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

X_train, X_test, y_train, y_test = train_test_split(np.array(X_pca), np.array(y), test_size=0.30, random_state=21)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
y_train = y_train.astype('float32')
y_test = y_test.astype('float32')

# Build the model architecture
model = Sequential([
    Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')  # sigmoid for binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train,
                    epochs=100,               # Number of epochs
                    batch_size=32,           # Batch size
                    validation_split=0.3,    # Validation split from training data
                    verbose=1)

# Evaluate the model on the test set
from sklearn.metrics import accuracy_score
test_loss, test_accuracy = model.evaluate(X_test, y_test)

print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
y_pred = (model.predict(X_test) > 0.5).astype("int64")
print("Accuracy Score on Test Data:", accuracy_score(y_test, y_pred))

from sklearn.metrics import classification_report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
